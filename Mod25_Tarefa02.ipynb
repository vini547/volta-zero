```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
from IPython.display import HTML
import warnings
```

```python
# Definir a função x^3
def f(x):
    return x**3
```

A função \( f(x) = x^3 \) é definida. Esta é a função que desejamos integrar numericamente, ou seja, a função que descreve a curva cujos valores de \( x \) (o domínio) mapeiam-se para \( f(x) \) (o contradomínio).



```python
a, b = 1, 2  # Intervalo de integração
```

Aqui, \( a = 1 \) e \( b = 2 \) representam os limites do intervalo de integração. Isso significa que estamos interessados em calcular a integral definida no intervalo de \( x = 1 \) a \( x = 2 \). Matematicamente, estamos calculando:
\[
\int_1^2 x^3 \, dx
\]


```python
n_max = 100  # Número máximo de retângulos para animação
```

Aqui, \( n_{\text{max}} = 100 \) define o número máximo de retângulos que serão usados para aproximar a integral pela soma de Riemann. Quanto maior o número de retângulos \( n \), mais precisa será a aproximação da área sob a curva.


```python
x_vals = np.linspace(a, b, 1000)  # Para desenhar a curva
```

A função `np.linspace(a, b, 1000)` cria uma sequência de 1000 valores uniformemente espaçados entre \( a = 1 \) e \( b = 2 \). Esses valores de \( x \) serão usados para traçar a curva contínua da função \( f(x) = x^3 \).


```python
fig, ax = plt.subplots()
```

`plt.subplots()` cria a estrutura para a figura e os eixos onde o gráfico será desenhado. Isso nos permite desenhar tanto a curva contínua quanto os retângulos da soma de Riemann.
    

```python
ax.plot(x_vals, f(x_vals), color='blue')  # Plot da curva f(x) = x^3
```

Esta linha traça o gráfico da função \( f(x) = x^3 \) usando os valores de \( x \) gerados anteriormente (com 1000 pontos). O gráfico da curva azul será a referência visual para a área sob a curva que estamos tentando aproximar.


```python
rects = ax.bar([], [], width=0, align='edge', color='orange', edgecolor='black', alpha=0.7)
```

Aqui, estamos preparando a estrutura para desenhar os retângulos da soma de Riemann. Inicialmente, os retângulos não têm largura (por isso a largura é zero). Eles serão atualizados durante a animação para representar a aproximação da integral.


```python
ax.set_xlim([a, b])
ax.set_ylim([0, 10])
ax.set_xlabel('x')
ax.set_ylabel('f(x) = x^3')
ax.set_title('Soma de Riemann: Aproximação da Integral')
```

Estas linhas configuram o gráfico ajustando os limites dos eixos \( x \) e \( y \). O eixo \( x \) é configurado para mostrar o intervalo de \( a = 1 \) até \( b = 2 \), e o eixo \( y \) é limitado para exibir valores entre 0 e 10 (um intervalo apropriado para a função \( f(x) = x^3 \) no intervalo de \( x \) que estamos analisando). Os rótulos dos eixos e o título do gráfico também são definidos aqui.


```python
def update(n):
    ax.clear()
    x = np.linspace(a, b, n+1)
    dx = (b - a) / n
    x_mid = x[:-1] + dx / 2  # Ponto médio para Riemann
```

Esta é a função de atualização da animação:
- `x = np.linspace(a, b, n+1)` divide o intervalo de \( a \) a \( b \) em \( n \) subintervalos.
- `dx = (b - a) / n` calcula a largura de cada subintervalo \( dx \).
- `x_mid = x[:-1] + dx / 2` calcula o **ponto médio** de cada subintervalo, que é usado na soma de Riemann com ponto médio.


```python
ax.plot(x_vals, f(x_vals), color='blue')
ax.set_xlim([a, b])
ax.set_ylim([0, 10])
ax.set_xlabel('x')
ax.set_ylabel('f(x) = x^3')
ax.set_title(f'Soma de Riemann com {n} retângulos')
```

Estas linhas redesenham a curva \( f(x) = x^3 \) e ajustam os limites dos eixos, os rótulos e o título do gráfico para cada atualização da animação, mantendo o gráfico atualizado conforme o número de retângulos \( n \) aumenta.


```python
ax.bar(x[:-1], f(x_mid), width=dx, align='edge', color='orange', edgecolor='black', alpha=0.7)
```

Esta linha desenha os retângulos da soma de Riemann. Para cada subintervalo:
- A base de cada retângulo está em \( x[:-1] \),
- A altura de cada retângulo é \( f(x_{\text{mid}}) \), que é o valor da função no ponto médio do subintervalo,
- A largura de cada retângulo é \( dx \).
Os retângulos representam a aproximação da área sob a curva usando a soma de Riemann com ponto médio.


```python
ani = FuncAnimation(fig, update, frames=np.arange(2, n_max+1), repeat=False)
```

Aqui, a função `FuncAnimation` cria a animação que atualiza o gráfico para cada valor de \( n \), variando o número de retângulos de 2 até \( n_{\text{max}} = 100 \). A função `update(n)` é chamada a cada frame da animação.


```python
HTML(ani.to_jshtml())
```

Esta linha converte a animação para um formato HTML que pode ser exibido diretamente no Jupyter Notebook.


# Explicação da Integral Analítica:
Para calcular a integral da função \( f(x) = x^3 \) entre os limites \( a = 1 \) e \( b = 2 \), seguimos os seguintes passos:

1. **Definição da Integral**:
   Queremos calcular a integral definida:
   \[
   \int_1^2 x^3 \, dx
   \]

2. **Cálculo da Antiderivada**:
   Primeiro, encontramos a antiderivada de \( f(x) = x^3 \):
   \[
   F(x) = \frac{x^4}{4} + C
   \]
   onde \( C \) é uma constante de integração que não influencia a integral definida.

3. **Avaliação nos Limites**:
   Em seguida, avaliamos \( F(x) \) nos limites superior e inferior:
   \[
   F(2) = \frac{2^4}{4} = \frac{16}{4} = 4
   \]
   \[
   F(1) = \frac{1^4}{4} = \frac{1}{4}
   \]

4. **Subtração dos Resultados**:
   Finalmente, subtraímos o valor da antiderivada no limite inferior do limite superior:
   \[
   \int_1^2 x^3 \, dx = F(2) - F(1) = 4 - \frac{1}{4} = \frac{16}{4} - \frac{1}{4} = \frac{15}{4}
   \]

5. **Resultado Final**:
   Portanto, o valor da integral definida é:
   \[
   \int_1^2 x^3 \, dx = \frac{15}{4} = 3.75
   \]
