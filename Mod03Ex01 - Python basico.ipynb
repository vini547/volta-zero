### 05 - Desafio
O algorítmo do exercício anterior não é o mais eficiente. O que você pode fazer para deixá-lo mais eficiente? Ou seja, executar menos comparações, portanto consumir menos tempo.
1. Será que precisamos correr o loop até o final sempre?
2. Será que precisamos mesmo verificar **todos** os números?
3. Será que precisamos ir até N-1?

Essas perguntas levam ao tipo de pensamento voltado a deixar um algoritmo mais eficiente. Veja se você consegue melhorar o seu.

````python
N = 98

# seu código aqui

def is_prime1(x):
    if x <= 1:
        print(f'{x} não é primo')
        return False
    
    if x <= 3:
        print(f'{x} é primo')
        return True
    
    if x % 2 == 0 or x % 3 == 0:
        print(f'{x} não é primo')
        return False   
    
    i = 5
    while i * i <= x:
        if x % i == 0 or x % (i + 2) == 0:
            print(f'{x} não é primo')
            return False
        i += 6
    print(f'{x} é primo')
    return True

def is_prime2(x):
    buff = []
    for i in range(1,x):
        buff.append(x%i)

    if buff.count(0) > 2:
        print(f'{x} não é primo')
    else:
        print(f'{x} é primo')
    return buff

t0 = time.time()
is_prime1(N)
t1 = time.time()
elapsed1 = t1 - t0
print(f'{t1-t0} seconds')

t0_2 = time.time()
is_prime2(N)
t1_2 = time.time()
elapsed2 = t1_2 - t0_2
print(f'{t1_2 - t0_2} seconds')



````

Para melhorar a performance deste algoritmo podemos utilizar atalhos que irão minimizar o número de iterações para se descobrir se um número é primo ou não. Para isso, a lista de premissas a seguir foi implementada no algoritmo:
1. Se um número não é primo, ele pode ser escrito na forma x = a + b. No caso de a e b serem maiores que a raíz quadrada de x, então o produto a x b também é maior que x. Desta forma, pelo menos a ou b deve ser menor ou igual à raiz quadrada de x.
2. Números menores ou iguals a 1 não são primos.
3. Números divisíveis por 2 e 3 não são primos.
4. Números pares maiores que 2 não são primos.
5. Qualquer número primo deve ser ímpar.
6. Qualquer número inteiro pode ser descrito como nas formas a seguir baseadas no srestos das divisões por 6 afim de se eliminar múltiplos de 2 e 3: 6k, 6k +1, 6k +2, 6k +3 etc..

### 06 - Peso ideal 1
O IMC (índice de massa corpórea) é um indicador de saúde mais bem aceito que o peso. Ele é calculado como:

$$ IMC = \dfrac{peso}{altura^2}$$

Segundo a OMS, valores *normais* são entre 18.5 e 24.9.

Sua tarefa é encontrar o ponto médio dessa faixa.

````python
import numpy as np
imc_ideal = np.arange(18.5, 24.9, 0.5).mean()
imc_ideal
````

### 07 - Peso ideal 2
Recebendo um valor de altura, encontre o peso '*ideal*' dessa pessoa, que fornece o IMC encontrado acima

````python
altura = 1.70

# Seu código
peso_ideal = round(imc_ideal * altura**2,2)
peso_ideal
````

### 08 - Peso ideal 3
Dada uma lista contendo as alturas de pacientes, crie uma nova lista que contenha o peso '*ideal*' (que fornece o IMC calculado em **Peso ideal 1**) desses pacientes.

````python
lista_alturas = [1.95, 2.05, 1.70, 1.65]

lista_peso_ideal = []

# seu código
for i in lista_alturas:
    lista_peso_ideal.append(round(imc_ideal * i**2,2))


lista_peso_ideal
````
