```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
df = pd.read_csv('./SINASC_RO_2019.csv',na_values='NA')
print(df.shape)
```

```python
sns.boxplot(
            y=df['IDADEMAE'],
            data=df,
            )
```

####  2) Faça um boxplot do peso do bebê

```python
sns.boxplot(
            y=df['PESO'],
            data=df,
            )
```

####  3) Faça um histograma da idade da mãe

```python
sns.displot(x=df['IDADEMAE'],
            data=df,
            alpha=0.25,
            bins=20,
            element='step',
            kde=True)
```

#### 4) Faça um histograma do peso do bebê

```python
sns.displot(x=df['PESO'],
            data=df,
            alpha=0.25,
            bins=20,
            element='step',
            kde=True)
```
