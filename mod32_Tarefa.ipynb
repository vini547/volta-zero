```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

1. Uma mquina embala rao para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatrio, com distribuio Normal, com mdia 15kg, mas com desvio padro 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produo tenha mais que 15,2 kg.

### Clculo da Probabilidade

Uma mquina embala rao para cachorro em pacotes de 15 kg. O peso dos pacotes segue uma **distribuio normal** com:


- Mdia $ \mu = 15 \, \text{kg} $  
- Desvio padro $ \sigma = 0,1 \, \text{kg} $
 


Queremos encontrar a probabilidade de que um pacote selecionado aleatoriamente pese **mais que 15,2 kg**, ou seja, \( P(X > 15,2) \).

#### 1. Padronizando a Varivel Aleatria

A frmula de padronizao :

$$
Z = \frac{X - \mu}{\sigma}
$$

Substituindo os valores \( X = 15,2 \), \( \mu = 15 \) e \( \sigma = 0,1 \):

$$
Z = \frac{15,2 - 15}{0,1} = \frac{0,2}{0,1} = 2
$$

#### 2. Calculando a Probabilidade

Queremos a probabilidade associada a \( Z > 2 \). Pela tabela da distribuio normal padro:

$$
P(Z > 2) = 1 - P(Z \leq 2)
$$

Pela tabela, \( P(Z \leq 2) = 0,9772 \). Assim:

$$
P(Z > 2) = 1 - 0,9772 = 0,0228
$$

#### 3. Resposta Final

A probabilidade de que um pacote pese mais que 15,2 kg :

$$
P(X > 15,2) = 0,0228 \quad (2,28\%)
$$


```python
mu = 15  # mdia
sigma = 0.1  # desvio padro
x = 15.2  # valor alvo

# Calculando o valor Z
z = (x - mu) / sigma
print(f"Valor de Z: {z}")

# Calculando a probabilidade P(X > 15.2)
prob = 1 - stats.norm.cdf(z)
print(f"Probabilidade P(X > 15.2): {prob:.5f} ({prob * 100:.2f}%)")
```

2. Da mquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
mu = 15  # mdia
sigma = 0.1  # desvio padro

percentil_95 = stats.norm.ppf(0.95, loc=mu, scale=sigma)
print(f"Percentil 95%: {percentil_95:.2f} kg")
```

3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na clula abaixo:

```python

Y = np.linspace(-6, 6, 200)

# Calculando as probabilidades para a distribuio Normal Padro (mdia=0, desvio padro=1)
prob_normal = stats.norm.cdf(Y, loc=0, scale=1)

# Calculando as probabilidades para a distribuio t-Student com 5 graus de liberdade
prob_t_student = stats.t.cdf(Y, df=5)

# Exibindo os primeiros 10 resultados para verificao
for y, p_norm, p_t in zip(Y[:10], prob_normal[:10], prob_t_student[:10]):
    print(f"Y = {y:.2f} | P_Norm(Y <= {y:.2f}) = {p_norm:.8f} | P_t(Y <= {y:.2f}) = {p_t:.8f}")

```

4. Faa uma visualizao grfica dos dados do item anterior.

```python
# Criando a visualizao grfica
plt.figure(figsize=(12, 6))

# Plotando a distribuio Normal Padro
plt.plot(Y, prob_normal, label='Normal Padro', color='blue', linewidth=2)

# Plotando a distribuio t-Student com 5 graus de liberdade
plt.plot(Y, prob_t_student, label='t-Student (df=5)', color='orange', linewidth=2)

# Adicionando ttulo e rtulos
plt.title('Distribuies Acumuladas: Normal Padro e t-Student')
plt.xlabel('Valores de Y')
plt.ylabel('Probabilidade Acumulada')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.grid()

# Adicionando legenda
plt.legend()
plt.xlim(-6, 6)
plt.ylim(-0.05, 1.05)

# Exibindo o grfico
plt.show()
```

5. 3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
# Definindo o objeto P com 20 valores igualmente espaados entre 0 e 1
P = np.linspace(0, 1, 20)

# Calculando os percentis para a distribuio Normal Padro (mdia=0, desvio padro=1)
percentis_normal = stats.norm.ppf(P, loc=0, scale=1)

# Calculando os percentis para a distribuio t-Student com 5 graus de liberdade
percentis_t_student = stats.t.ppf(P, df=5)

# Exibindo os resultados
print("Percentis para a distribuio Normal Padro:")
for p, q in zip(P, percentis_normal):
    print(f"P = {p:.2f} | Percentil = {q:.5f}")

print("\nPercentis para a distribuio t-Student (5 graus de liberdade):")
for p, q in zip(P, percentis_t_student):
    print(f"P = {p:.2f} | Percentil = {q:.5f}")
```

6. Faa um grfico com os dados do item anterior.

```python
plt.figure(figsize=(12, 6))

# Plotando percentis da distribuio Normal Padro
plt.plot(P, percentis_normal, label='Percentis Normal Padro', marker='o', color='blue')

# Plotando percentis da distribuio t-Student
plt.plot(P, percentis_t_student, label='Percentis t-Student (df=5)', marker='o', color='orange')

# Adicionando ttulo e rtulos
plt.title('Percentis das Distribuies Normal Padro e t-Student')
plt.xlabel('Valores de P')
plt.ylabel('Percentil')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.grid()

# Adicionando legenda
plt.legend()
plt.ylim(-5, 5)

# Exibindo o grfico
plt.show()
```
