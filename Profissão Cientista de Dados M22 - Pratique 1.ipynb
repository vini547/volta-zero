#### 1. Baixando dados de ativos
    1. Baixe os dados de três ativos da sua escolha do Yahoo Finance, correspondente a um período de pelo menos 2 anos até a data mais recente.
    2. Salve esses dados para que você possa utilizá-los na próxima tarefa.

```python
import yfinance as yf
import pandas as pd
import plotly.express as px
import datetime

# Definindo intervalo de datas
start_date = "2023-07-01"
end_date = datetime.datetime.today().strftime('%Y-%m-%d')

# Lista de ativos escolhidos
ativos = ['AAPL', 'MSFT', 'GOOGL']
dados = {}

# Baixando e salvando os dados
for ativo in ativos:
    df = yf.download(ativo, start=start_date, end=end_date)
    df['Amplitude'] = df['High'] - df['Low']  # Amplitude diária
    df['Ativo'] = ativo
    dados[ativo] = df
    df.to_csv(f"{ativo}_historico.csv")  # Salvando CSV para uso posterior

# Concatenando dados em um único DataFrame
df_amplitude = pd.concat([dados[ativo][['Amplitude', 'Ativo']] for ativo in ativos])
df_amplitude['Data'] = df_amplitude.index

# Gráfico da amplitude diária
fig = px.line(df_amplitude, x='Data', y='Amplitude', color='Ativo',
              title='Amplitude diária (volatilidade intradiária) dos ativos')
fig.show()

# Verificando qual ativo foi mais volátil (maior amplitude média)
media_amplitude = df_amplitude.groupby('Ativo')['Amplitude'].mean().sort_values(ascending=False)
print("Média da amplitude intradiária por ativo:\n")
print(media_amplitude)
```
