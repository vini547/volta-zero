```python
import yfinance as yf
import pandas as pd
import plotly.express as px
import datetime
import plotly.graph_objects as go





# Definindo intervalo de datas
start_date = "2023-07-01"
end_date = datetime.datetime.today().strftime('%Y-%m-%d')

# Lista de ativos escolhidos
ativos = ['AAPL', 'MSFT', 'GOOGL']
dados = {}

# Baixando e salvando os dados
for ativo in ativos:
    df = yf.download(ativo, start=start_date, end=end_date)
    df['Amplitude'] = df['High'] - df['Low']  # Amplitude diária
    df['Ativo'] = ativo
    dados[ativo] = df
    df.to_csv(f"{ativo}_historico.csv")  # Salvando CSV para uso posterior

# Concatenando dados em um único DataFrame
df_amplitude = pd.concat([dados[ativo][['Amplitude', 'Ativo']] for ativo in ativos])
df_amplitude['Data'] = df_amplitude.index

# Gráfico da amplitude diária
fig = px.line(df_amplitude, x='Data', y='Amplitude', color='Ativo',
              title='Amplitude diária (volatilidade intradiária) dos ativos')
fig.show()

# Verificando qual ativo foi mais volátil (maior amplitude média)
media_amplitude = df_amplitude.groupby('Ativo')['Amplitude'].mean().sort_values(ascending=False)
print("Média da amplitude intradiária por ativo:\n")
print(media_amplitude)
```

```python
# Lista dos ativos
ativos = ['AAPL', 'MSFT', 'GOOGL']
dados = {}

# Carregando os arquivos CSV salvos anteriormente
dados = {}
ativos = ['AAPL', 'MSFT', 'GOOGL']

for ativo in ativos:
    df = pd.read_csv(f"{ativo}_historico.csv", index_col=0, parse_dates=True)

    # Remove possíveis linhas inválidas (como 'Ticker' no final)
    df = df[pd.to_numeric(df['Close'], errors='coerce').notnull()]

    # Converte colunas numéricas de object para float
    colunas_numericas = ['Close', 'High', 'Low', 'Open', 'Volume']
    for col in colunas_numericas:
        df[col] = pd.to_numeric(df[col], errors='coerce')

    # Ordena e salva
    df.sort_index(inplace=True)
    dados[ativo] = df

# Cálculo da média móvel e desvio padrão com janela de 90 dias
for ativo in ativos:
    dados[ativo]['Media_Movel_90'] = dados[ativo]['Close'].rolling(window=90).mean()
    dados[ativo]['Desvio_Padrao_90'] = dados[ativo]['Close'].rolling(window=90).std()
    dados[ativo]['Ativo'] = ativo

# Concatenando dados para facilitar os gráficos
df_media = pd.concat([dados[ativo][['Media_Movel_90', 'Ativo']] for ativo in ativos])
df_media['Data'] = df_media.index

df_std = pd.concat([dados[ativo][['Desvio_Padrao_90', 'Ativo']] for ativo in ativos])
df_std['Data'] = df_std.index

# Gráfico de linha: Média móvel de 90 dias
fig_media = px.line(df_media, x='Data', y='Media_Movel_90', color='Ativo',
                    title='Média Móvel de 90 dias (Adj Close)')
fig_media.show()

# Gráfico de linha: Desvio padrão móvel de 90 dias
fig_std = px.line(df_std, x='Data', y='Desvio_Padrao_90', color='Ativo',
                  title='Desvio Padrão Móvel de 90 dias (Adj Close)')
fig_std.show()

# Análise adicional: volatilidade média
print("\nVolatilidade média (desvio padrão móvel):")
print(df_std.groupby('Ativo')['Desvio_Padrao_90'].mean().sort_values(ascending=False))

# Gráfico de Candlestick dos últimos 60 dias para o ativo AAPL
ativo_candle = 'AAPL'
dados_candle = dados[ativo_candle].copy()

# Força o índice a ser datetime, caso ainda esteja como string
dados_candle.index = pd.to_datetime(dados_candle.index)

# Seleciona os últimos 60 dias
dados_candle_60dias = dados_candle.last('60D')

# Gera o gráfico
fig_candle = go.Figure(data=[go.Candlestick(
    x=dados_candle_60dias.index,
    open=dados_candle_60dias['Open'],
    high=dados_candle_60dias['High'],
    low=dados_candle_60dias['Low'],
    close=dados_candle_60dias['Close'],
    name=ativo_candle
)])
fig_candle.update_layout(title=f'Candlestick dos últimos 60 dias - {ativo_candle}',
                         xaxis_title='Data', yaxis_title='Preço')
fig_candle.show()
```
